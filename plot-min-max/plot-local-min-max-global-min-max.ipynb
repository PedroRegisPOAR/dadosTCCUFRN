```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t = np.arange(0.25, 2.25, 0.0001)


plt.ylim(-0.9, 0.7)
plt.plot(t, f(t), 'k', label=r"$f(x)=e^{-x}\!\cos\left ( 2\pi x \right )$")

plt.annotate('Máx. global', xy=(1, 0.38), xytext=(1.3, 0.45),
            arrowprops=dict(arrowstyle= '->', facecolor='black', lw=2))

plt.annotate('Máx. local', xy=(2, 0.15), xytext=(2, 0.3),
            arrowprops=dict(arrowstyle= '->', facecolor='black', lw=2))


plt.annotate('Mín. global', xy=(0.5, -0.65), xytext=(0.45, -0.85),
            arrowprops=dict(arrowstyle= '->', facecolor='black', lw=2))

plt.annotate('Mín. local', xy=(1.475, -0.25), xytext=(1.3, -0.5),
            arrowprops=dict(arrowstyle= '->', facecolor='black', lw=2))


plt.legend(loc='upper left')

plt.savefig('figura-local-min-max-global-min-max.pdf', dpi = 300)

plt.show()
# f\!\left ( x \right )=e^{-x}\!\cos\left ( 2\pi x \right )
```
